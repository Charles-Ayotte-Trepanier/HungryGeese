```python
import pickle
from agents.nnAgent import nnModel, train_test_splitter
import tensorflow as tf

tf.compat.v1.reset_default_graph()

policy_model = nnModel('policy')
policy_model.build_model()

v_model = nnModel('state_value')
v_model.build_model()
```

```python
nb_passes = 1
nb_files = 41

for it in range(nb_passes):
    print(f'Pass #{it}')
    for i in range(nb_files):
        path = f'./data/it_{i}.pkl'
        with open(path, 'rb') as f:
            data = pickle.load(f)
        print(f'learning using "{path}"')
        X = data['state']
        y = data['y']
        v = data['v']
        X_train, X_test, y_train, y_test, v_train, v_test = train_test_splitter(X, y, 0.05, v=v)
        
        policy_model.fit(X_train,
                         y_train,
                         X_test,
                         y_test,
                         epoch=5,
                         batch_size=1024)
        
policy_model.save('rule_based_policy')

v_model.replace_weights(policy_model.model)
for it in range(nb_passes):
    print(f'Pass #{it}')
    for i in range(nb_files):
        path = f'./data/it_{i}.pkl'
        with open(path, 'rb') as f:
            data = pickle.load(f)
        print(f'learning using "{path}"')
        X = data['state']
        y = data['y']
        v = data['v']
        X_train, X_test, y_train, y_test, v_train, v_test = train_test_splitter(X, y, 0.05, v=v)

        
        v_model.replace_weights(policy_model.model)
        v_model.fit(X_train,
                    v_train,
                    X_test,
                    v_test,
                    epoch=5,
                    batch_size=1024)
```
